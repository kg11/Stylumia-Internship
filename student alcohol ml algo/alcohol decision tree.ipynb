```r
library("caTools")
library(MASS)
library(rpart)
library(party)
```

```r
str(d1)
col=c("Medu","Fedu","traveltime","studytime","failures","famrel","freetime","goout","Dalc","Walc","health")
```

```r
#converting categorical values as factor
for(i in col)
{
  d1[,i]=as.factor(d1[,i])
}
summary(d1)
```

```r
#i am comsidering avgmarks as new attribute(g1+g2+g3)/3 and removing g1,g2,g3 and only concidering avggrades
d1$avggrades=(d1$G1+d1$G2+d1$G3)/3
d1=d1[,c(1:30,34)]
```

```r
str(d1)
```

```r
#now splitting dataset into two parts train and test for decision tree
ind=sample.split(Y=d1$avggrades,SplitRatio = 0.8)
train=d1[ind,]
test=d1[!ind,]
```

```r
#using ctree to make decision tree
pct=ctree(avggrades~ . ,data=train)
plot(pct)
```

```r
#now predicting on test data
pwc=predict(pct,test)
pwc
```

```r
#finding accuracy
nmse.lm<-mean((pwc-test[,"avggrades"])^2)/mean((mean(test$avggrades)-test[,"avggrades"])^2)
print(nmse.lm)
```

```r
#so we see that our model is preducting with 86% accuracy which is great
```
